```python
ruta ='/content/drive/MyDrive/DATASIA/SiaCiruela_Durazno.xlsx'
```

```python
import pandas as pd

# Leer la primer hoja del Excel
df = pd.read_excel(ruta)

# Ver las primeras filas
df.head()
```

```python
# Variables en el tiene el archivo
df.columns
```

```python
#Una primera aproximación para entender nuestra base (por lo menos los valores numéricos) se usa describe()
print(df.describe())
```

```python
# Selecciono un cultivo
# Filtrar las filas donde la columna 'Cultivo' sea igual a Vid'
uva = df[df['Cultivotipo'] == 'Vid']
uva.head()
```

```python
superficie_por_anio = uva.groupby('Anio')['SuperficieLote'].mean().reset_index()
superficie_por_anio.head()
```

```python
# @title SuperficieLote

from matplotlib import pyplot as plt
superficie_por_anio['SuperficieLote'].plot(kind='line', figsize=(8, 4), title='SuperficieLote')
plt.gca().spines[['top', 'right']].set_visible(False)
```
